```python
categorical_columns = df.select_dtypes(include=['str'])
numerical_columns = df.select_dtypes(exclude=['str'])
```

```python
#IN CASE A NULL VALUE IS PRESENT.
from pandas.api.types import is_numeric_dtype
#for numerical columns, place mean values at empty records.
#for categorical columns, place modal value at  empty records.

for i in df.columns:
    if is_numeric_dtype(df[i]):
        mean = df[i].mean()
        df[i] = df[i].fillna(mean)
    else:
        mode = df[i].mode()[0]
        df[i] = df[i].fillna(mode)
```

```python
unusual_amounts = df['unusual_amount_flag'].sum()
print("Accounts with unusual amounts, than the usual balance:", unusual_amounts)
```

```python
# In a fraudulent transaction dataset, a velocity_flag (or velocity feature) signifies the frequency, speed, or 
# volume of transactions associated with a specific entity (card, user, IP address, or device) within a defined,
# short time frame. It is designed to catch fraudsters who attempt to maximize profit from a stolen card or
# account before it is blocked

velocity_flags = df['velocity_flag'].sum()
print("Velocity flagged transactions:", velocity_flags)
```
